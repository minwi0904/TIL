```python
import pandas as pd
import seaborn as sns
```

```python
body_df = pd.read_csv('body.csv', index_col=0)
body_df.head()
```

```python
body_df['Height']
```

```python
body_df['Height'].value_counts().sort_index() # sort_index()하면 인덱스 기준으로 정렬됨
```

```python
body_df['Height'].value_counts().sort_index().plot()
```

```python
sns.kdeplot(body_df['Height'], bw =0.5) # 여기서 씨본사용, bw는 히스토그램에서 구간 나눠주는거랑 비슷해
```

```python
body_df.plot(kind='hist', y='Height', bins=15)
```

```python
sns.distplot(body_df['Height'], bins=15)
```

```python
body_df.plot(kind='box', y='Height')
```

```python
sns.violinplot(y=body_df['Height'])
```

```python
body_df.plot(kind='scatter', x='Height', y='Weight')
```

```python
sns.kdeplot(body_df['Height'], body_df['Weight']) # 등고선 같아 바깥으로갈수록 내려가,
# 아래 키와 몸무게 그래프 합쳐서 3차원, 선끼리 떨어져있으면 완만, 붙어있으면 가팔라!
```

```python
sns.kdeplot(body_df['Height']) # 키
```

```python
sns.kdeplot(body_df['Weight']) #몸무게
```

```python
sns.lmplot(data=body_df, x='Height', y='Weight') # 회귀선을 그려줘 추세선같은 느낌인가
```

```python
laptops_df = pd.read_csv('laptops.csv')
laptops_df.head()
```

```python
laptops_df['os'].unique()
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='box')
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='violin')
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='strip')
```

```python
laptops_df['processor_brand'].unique()
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='strip', hue='processor_brand')
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='swarm', hue='processor_brand')
```
