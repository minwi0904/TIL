```python
laptop = pd.read_csv('laptops.csv')
laptop.head()
```

```python
brand_nation = {
    'Dell': 'U.S.',
    'Apple': 'U.S.',
    'Acer': 'Taiwan',
    'HP': 'U.S.',
    'Lenovo': 'China',
    'Alienware': 'U.S.',
    'Microsoft': 'U.S.',
    'Asus': 'Taiwan',
}
```

```python
laptop['brand'].map(brand_nation)
```

```python
laptop['brand_nation'] = laptop['brand'].map(brand_nation)
```

```python
laptop
```

```python
nation_groups = laptop.groupby('brand_nation')
type(nation_groups)
```

```python
nation_groups.count()
```

```python
nation_groups.max()
```

```python
nation_groups.mean()
```

```python
nation_groups.first()
```

```python
nation_groups.last()
```

```python
nation_groups.plot(kind='box', y='price')
```

```python
nation_groups.plot(kind='hist', y='price')
```
