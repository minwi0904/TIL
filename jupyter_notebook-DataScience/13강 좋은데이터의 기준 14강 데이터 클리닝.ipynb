## 좋은 데이터 기준 = 완결성, 유일성, 통일성, 정확성
#### 완결성 = 필요한값이 전부 있냐 = 결측값이 없다
#### 유일성 = 동일한 값이 한가지만 있냐
#### 통일성 = 동일한 형식으로 저장이 되어있냐
#### 정확성 = 데이터가 정확하게 모아졌냐(아웃라이어가 없으면 좋아)

절대적인 기준은 아니고 기본설정임

```python
beer = pd.read_csv('beer.csv', index_col=0)
beer.head()
```

```python
beer.plot(kind='box', y='abv')
```

```python
beer['abv'].describe()
```

```python
beer['abv'].quantile(0.25)
```

```python
beer['abv'].quantile(0.75)
```

```python
q1 = beer['abv'].quantile(0.25)
q3 = beer['abv'].quantile(0.75)
iqr = q3 - q1
```

```python
condition = (beer['abv'] < q1 - 1.5 * iqr) | (beer['abv'] > q3 + 1.5 * iqr)
```

```python
beer[condition]
```

```python
beer.loc[2250, 'abv'] = 0.055
beer.loc[2250]
```

```python
condition = (beer['abv'] < q1 - 1.5 * iqr) | (beer['abv'] > q3 + 1.5 * iqr)
beer[condition]
```

```python
beer[condition].index
```

```python
beer.drop(beer[condition].index, inplace=True)
```

```python
condition = (beer['abv'] < q1 - 1.5 * iqr) | (beer['abv'] > q3 + 1.5 * iqr)
beer[condition]
```

```python
beer.plot(kind='box', y='abv')
```
