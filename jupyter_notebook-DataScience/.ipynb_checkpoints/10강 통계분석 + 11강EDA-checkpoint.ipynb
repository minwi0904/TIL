```python
클러스터 분석(cluster analysis) : 데이터를 몇가지 무리로 나눠서 분석 느낌
```

```python
interests = df.loc[:, 'History':'Pets']
interests.head()
```

```python
corr = interests.corr()
corr
```

```python
corr['History'].sort_values(ascending=False)
```

```python
sns.clustermap(corr) # 히트맵이랑 비슷한데 위에 사다리처럼 연결된 그림이 연관된걸 묶어놓은것!
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.head()
```

```python
titacorr = titanic.corr()
titacorr
```

```python
titanic['Survived'].value_counts() # 사망자가 더 많다 3번
```

```python
titanic['Pclass'].value_counts()# 3등실에 가장 많은 사람이 탔다 4번
```

```python
titanic['Age'].plot(kind='hist', bins=30) 
# 나이분포 승객은 20대, 30대가 가장 많다. 1번
```

```python
sns.catplot(data=titanic, x='Survived', y='Age', kind='swarm')
# 나이가 어릴수록 생존율이 높지 않다. 6번
# 높다면 두개를 비교했을때주황색쪽이 파란색에 비해 나이가 줄어들수로 ㄱ커져야함
```

```python
# 나이별 요금 분포
titanic.plot(kind='scatter', x='Age', y='Fare')
```

```python
# 객실 등급별 생존율 분포 >> 
# 산점도를 사용하면 됨, 근데 값이 단일 값이라 보기어려우니 KEDplot활용
# 이렇게 하면 빈도도 알수 있어
sns.kdeplot(titanic['Pclass'], titanic['Survived'])
# 확실히 3등급 사람들끼리 비교했을때 대다수가 아래에 밀집됐고,
# 2등실은 비슷한 반면
# 1등실의 살아있는사람비중이 가장 큼
```

```python
# 생존여부에 따른 나이 및 성별 분포
sns.stripplot(data=titanic, x='Survived', y='Age', hue='Sex')
# 나이여부는 생존율에 영향을 주지 않았지만 확실히 여성이 더 많이 살았다.
```
